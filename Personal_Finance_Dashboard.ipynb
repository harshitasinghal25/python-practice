```python
# Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sample finance data
Data = {
    'Age': [25, 30, 35, 40, 45, 50, 55, 60, 65, 70],
    'Income': [50000, 60000, 70000, 80000, 90000, 100000, 110000, 120000, 130000, 140000],
    'Expenses': [20000, 25000, 30000, 35000, 40000, 45000, 50000, 55000, 60000, 65000],
}
df = pd.DataFrame(Data)
df['Savings'] = df['Income'] - df['Expenses']
df.head()
df.describe()
```

```python
# Shape of Data
print ('Shape:', df.shape)
```

```python
# Checking Missing Values
df.isnull().sum()
```

```python
# Visualisation
# Income vs Expense Line Chart
plt.figure(figsize=(10,6))
plt.plot(df['Age'], df['Income'], label='Income')
plt.plot(df['Age'], df['Expenses'], label='Expenses')
plt.title('Income vs Expenses')
plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
plt.show()
```

```python
# Visualisation
# Savings by Age bar chart
plt.figure(figsize=(10,6))
plt.bar(df['Age'], df['Savings'], label='Savings')
plt.title('Savings By Age')
plt.xlabel('Age')
plt.ylabel('Savings')
plt.legend()
plt.show()
```

```python
# Visualisation
# Savings Rate calculation and visualization
df['Savings_Rate'] = (df['Savings'] / df['Income']) * 100
plt.figure(figsize=(10,6))
plt.bar(df['Age'], df['Savings_Rate'], label='Savings Rate')
plt.title('Savings Rate By Age')
plt.xlabel('Age')
plt.ylabel('Savings Rate')
plt.legend()
plt.show()
```
